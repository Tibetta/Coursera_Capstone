# Segmenting and Clustering Neighborhoods in Toronto

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)
```

Scraping the data from the Wikipedia page using pandas

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

dfs = pd.read_html(url)
print(len(dfs))
```

In the web page there are 3 tables. I'll select the one with the Toronto data

```python
df = dfs[0]
df.head()
```

Ignore cells with a borough that is "Not assigned".

```python
df['Borough'].replace('Not assigned', np.nan, inplace=True)
df.dropna(subset=['Borough'], inplace=True)

df
```

Group by the Postal Code and reset the index


```python
df.groupby('Postal Code')
df_set = df.reset_index(drop=True)
df_set.head(20)
```

 If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.
 1. Check if there are 'Not assigned' Neighborhood in the dataframe
 2. replace the 'Not assigned' Neighborhood with the Borough

```python
x = len(df_set)
#print(x)
i = 0
word = 'Not assigned'
word
for i in range(x):
    #print(i)
    if df_set['Neighbourhood'].any() == word:
        df_set['Neighbourhood'] == df_set['Borough']
        print(df_set['Neighbourhood'])
    i = i + 1
print('I have finished to check the data frame')

```
